```python
import pandas as pd
import numpy as np
```

```python
tabuleiro = pd.DataFrame(index=range(1, 9), columns=range(1, 9)).fillna(0)
```

Função que irá detectar onde se encontram os 1 na array e verificar para cada 1 encontrado se mais alguns 1's em suas diagonais principal e secundaria, na coluna e fileira.

Para fazer a checagem pelas diagonais, será a partir da coordenada do 1 encontrado, descendo pela diagonal esquerda com loop -1 col, +1 row(seria essa a diagonal secundaria), após encontrado a última posição dessa diagonal é feito uma subida com o loop checando se há mais algum 1. O mesmo para direita ocorrendo a partir da coordenada do 1, incrementado +1 em row e col até chegar no último "do lado direito", e depois é feito um loop baixo para cima checando a diagonal dir (diagonal primaria)

```python
# Em python o primeiro elemento é da coluna o segunda da fileira
def queens(tabela):
    for icol in range(1, 9):

        for irow in range(1, 9):
            
            # -1 pois iremos ignorar a contagem do encontrado
            # Contagem de 1s na fileira
            fil = -1
            # Contagem de 1s na coluna
            col = -1
            # Contagem de 1s na diagonal direita
            dig_dir = -1
            # Contagem de 1s na diagonal esquerda
            dig_esq = -1

            if(tabuleiro[icol][irow] == 1):

                # Contar 1s na coluna icol
                for k in range(1, 9):
                    if (tabuleiro[icol][k] == 1):
                        col=col+1

                # Contar 1s na fileira irow
                for k in range(1, 9):
                    if (tabuleiro[k][irow] == 1):
                        fil=fil+1

                # encontrar diagonal esquerda (secundaria)
                col_esq = icol
                row_esq = irow
                while(col_esq != 1 and row_esq != 8):
                    col_esq-=1
                    row_esq+=1       

                # cencontrar diagonal direita (primaria)
                col_dir = icol
                row_dir = irow
                while(col_dir != 8 and row_dir != 8):
                    col_dir+=1
                    row_dir+=1

                # contar 1's diagonal esquerda (secundaria)
                while(row_esq>0 and col_esq<9):

                    if (tabuleiro[col_esq][row_esq] == 1):
                        dig_esq+=1

                    col_esq+=1
                    row_esq-=1

                # contar 1's diagonal direita (primaria)
                while(col_dir>0 and row_dir>0):

                    if (tabuleiro[col_dir][row_dir] == 1):
                        dig_dir+=1
                        
                    col_dir-=1
                    row_dir-=1

                # printando resultado   
                print("Posiçao row =",irow)
                print("Posiçao col =",icol)
                print("1's na fileira =",fil)
                print("1's na coluna =",col)
                print("1's na dig direita (principar) =",dig_dir)
                print("1's na dig esquerda (secundaria) =",dig_esq,"\n")
```
